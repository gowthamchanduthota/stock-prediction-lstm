```python
closing_data = aapl_data[['Close']]
closing_data.describe
```

```python
max(closing_data.values), min(closing_data.values)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(closing_data)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data = []
y_data = []
for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])


import numpy as np

x_data = np.array(x_data)
y_data = np.array(y_data)
```

```python
len(x_data)
x_data.shape
```

```python
splitlen = int(len(x_data)*0.7)
X_train = x_data[:splitlen]
y_train = y_data[:splitlen]

X_test = x_data[splitlen:]
y_test = y_data[splitlen:]
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model1 = Sequential()
model1.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model1.add(LSTM(50, return_sequences=True))
model1.add(LSTM(50))
model1.add(Dense(1))
model1.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model1.summary()
```

```python
model1.fit(X_train, y_train, batch_size=64, epochs=100, validation_data=(X_test, y_test), verbose=1)
```

```python
predictions = model1.predict(X_test)
inv_predications = scaler.inverse_transform(predictions)
inv_y = scaler.inverse_transform(y_test)
for i in range(len(predictions)):
    print("actual: {}   -  {}".format(inv_y[i], inv_predications[i]))
```

```python
import math
from sklearn.metrics import mean_squared_error

math.sqrt(mean_squared_error(inv_y, inv_predications))
```

```python
data_plot = pd.DataFrame({
'original_test_data': inv_y.reshape(-1),
'predicted_test_data': inv_predications.reshape(-1)
}, index=aapl_data.index[splitlen + 100:])

data_plot.head()
```

```python
plot_graph((15, 5), data_plot, "Predictions")
```

```python
plot_graph((15, 5), pd.concat([closing_data[:splitlen+100], data_plot], axis=0), "Total")
```
